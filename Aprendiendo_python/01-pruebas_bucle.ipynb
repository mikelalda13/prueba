guarda en una lista los numeros primos de 0 a 100

```python
# coger todos los nmeros de 0 a 100
for num in range(100):
    print(num)
    # Para cada nmero miramos si las divisiones de los numeros del numero-1 al 1 (si incluir el 1)
    for div in range(num-1,1,-1):
        print('div:', div)
```

```python
numeros_primos = set()
for num in range(2,100,1):
    primo = True
    for div in range(num-1,1,-1):
        if num%div==0:
            # No es primo
            primo = False
            break
    if primo:
        numeros_primos.add(num)
numeros_primos
```

```python
numeros_primos = set()
for num in range(2,100,1):
    primo = True
    for div in range(2,num,1):
        if num%div==0:
            # No es primo
            primo = False
            break
    if primo:
        numeros_primos.add(num)
numeros_primos
```
